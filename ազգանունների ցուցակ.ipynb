```python
a = [element.split("-")[0].rstrip().lstrip() for element in lines]
```

```python
#a
```

```python
b = [element.split(',') for element in a if len(element)>1]
```

```python
c = []
b[1][0]
for element in b:
    for i in element:
        c.append(i.strip())

    
```

```python
#c
```

```python
list_of_last_names = []
for element in c:
    list_of_last_names.append(element.rstrip())
```
